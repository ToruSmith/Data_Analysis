## plotly+cufflinks
This introduction could let you know the basics of plotly+cufflinks 

cufflinks: One library could let you combine pandas with plotly. It means that when you have a dataframe, you could easily use iplot to plot a fantastic figure.

```python
#Data Science Import
import pandas as pd
import numpy as np

#Warning that could ignore return
import warnings
warnings.filterwarnings('ignore')
```

### Data 

```python
data=pd.read_csv('Lebron_James_regular_stats.csv',index_col='Date',parse_dates=True)
data['MP']=data['MP'].apply(lambda x:x[:2]).astype(np.float)
data.rename(columns={'\xa0.1':'WIN','+/-':'score'},inplace=True)
data['WIN']=data['WIN'].apply(lambda x:0 if 'L' in x else 1)
columns=['TRB','AST','STL','BLK','TOV','PF','PTS','score','WIN']
df=data[columns]
df.head()
```

```python
#可以製作互動式圖表的套件
import plotly.plotly as py
import plotly.graph_objs as go
from plotly.offline import iplot,init_notebook_mode

#為可以將pandas和plotly結合的套件
import cufflinks
cufflinks.go_offline(connected=True)
init_notebook_mode(connected=True)
```

## Histogram
Simply use one variable to plot a histogram plot. Using 'xTitle', 'yTitle' and 'title' to add what you want. 

```python
df['PTS'].iplot(kind='hist', #histogram plot
                xTitle='score',
                yTitle='count',
                title='Claps Distribution')
```

Plot the different variables at the same time

```python
df[['TRB','AST']].iplot(kind='hist',
                        xTitle='score',
                        yTitle='count(%)',
                        title='Claps Distribution',
                        histnorm='percent',  #from 'frequency' to 'percent'
                        linecolor='Black',
                        bins=30,   #This would only valid in 'histogram'
                        barmode='group', #Mode that display the bar. ex.group,stack,overlay
                        bargap=0.1)  #gap among bars
```

# Bar  

Here I would explain the go.layout.XAxis this function. If you want to describe the text in the x axis or y axis in the plotly+cufflink, you should add the parameter 'layout'. Ticktext is the text you want to change. Tickvals that is the parameter in the layout should be the index of the dataframe.

```python
df2=df[['PTS','TRB','AST']].resample('A-SEP').mean()
ticktext=['第%d季' %i for i in range(1,df2.shape[0]+1)]
df2.iplot(kind='bar',
          linecolor='Black',
          layout={'xaxis':go.layout.XAxis(       #I change the graph_objs to go
          ticktext=ticktext,            #XAxis is used to describe the text and values in the xticks.
          tickvals=df2.index,
          title='季'),
          'yaxis':go.layout.YAxis(title='季平均'),
          'title':'Year Average Points, Rebounds and Assists'})
```

# Scatter
These plot would contain different type of scatter plot. You should check the parameters in the iplot I used. Please be patient to understand why I used these here. You could be rewarded a lot.

```python
data_season=data.resample('A-SEP')
text1=[f'<span style="color:blue">{season.year}年:平均勝{x:.2f}分<span><br>'  \
       if x>0 else f'<span style="color:blue">{season.year}年:平均輸{x:.2f}分<span><br>' \
       for season,x in zip(data_season.size().index,data_season['score'].mean())]
data_season.sum().iplot(y=['WIN'],
                         mode='lines+markers+text', #Here text is an important word because after you called that, this would appear in the data points.
                         opacity=0.5,
                         layout={'xaxis':go.layout.XAxis(
                            ticktext=ticktext,
                            tickvals=data_season.asfreq().index,
                            title='date'),
                            'yaxis':go.layout.YAxis(
                            title='勝分'),
                        'title':'每年平均勝分'},
                         text=text1)  #Add text in the plot 
```

```python
data['WIN']=data['WIN'].apply(lambda x:'WIN' if x==1 else 'LOSE')
data.iplot(x='PTS',
              y='AST',
              categories='WIN',
              xTitle='Points',
              yTitle='Assists',
              opacity=0.8)
```

```python
data.iplot(x='PTS',
           y='AST',
           mode='markers',
           size=data['MP'],      #Depende on the value of the variable to determine the size
           categories='WIN',    #One variable would contain different categories, so let the plot present
           xTitle='PTS',
           yTitle='AST',
           title='PTS vs AST by Game Minutes')
```

This woud be the new for you. go.Scatter would like a data you would use to plot the figure. And then go.Figure could let you determine the other except for the data, such as x axis, y axis or the title of the plot. 

```python
data_go=[go.Scatter(x=df['PTS'],
					y=df['AST'],
					mode='markers',
					marker=dict(colorscale='Rainbow',
					 			size=df['TRB']*5,
					 			color=df['WIN'],
					 			sizemin=10,
					 			showscale=True,  #Whether you want to add a colorbar.
					 			line=dict(color='black',width=3)))]

figure=go.Figure(data=data_go,
				 layout=go.Layout(
				 		xaxis=dict(title='POINTS'),
				 		yaxis=dict(title='ASSISTS'),
				 		title='POINTS vs ASSISTS'))

iplot(figure)
```

```python
data.resample('M').mean().iplot(
        y=['FG%','3P%'],
        mode='lines+markers',
        vline=[dict(x=pd.to_datetime('2014-07-01'),    #Emphasize the line you want you or other people could know more easier.
                       color='red',
                       dash='dash',
                       width=10),
                 dict(x=pd.to_datetime('2010-07-01'),
                       color='red',
                       dash='dash',
                       width=3)],
        size=8,
        symbol=1,     #different symbols represent different markers
        xTitle='Date',
        yTitle='Read Time',
        title='FG% & 3P%')
```

```python
data.iplot(y=['FG%','3P%'],mode='lines+markers',
              vspan=[dict(x0='2015-10-01',       #fill in the span 
                               x1='2015-12-31',
                               color='green',fill='green',
                               opacity=0.2)],
             hline=[dict(y=i,color='red',dash='dash',width=1) for i in [0.2,0.4,0.6,0.8]],
             vline=[dict(x=i,color='blue',dash='dash',width=1) for i in pd.date_range('2015-09-01',periods=8,freq='M')],
             xrange=('2015-09-01','2016-03-01'),  #like plt.xlim()  Restrict the xaxis range.
             opacity=0.8,
             size=10,symbol=5,xTitle='FG%',yTitle='3P%')
```

## Pairgrid
This plot is like the sns.pairgrid. You should specified what plots you want to use in the diagram, upper or lower.

```python
import plotly.figure_factory as ff

df1=data.copy()
df1['WIN']=data['WIN'].apply(lambda x:'WIN' if x==1 else 'LOSE')

figure=ff.create_scatterplotmatrix(
        df1[['PTS','TRB','AST','WIN']],
        height=1000,width=1000,
        diag='histogram',
        index='WIN')

iplot(figure)
```

# Heatmap
Heatmap is always the great one to describe the correlation of the features in the dataframe.

```python
corr=df.corr()

figure=ff.create_annotated_heatmap(
        z=corr.values,
        x=list(corr.columns),
        y=list(corr.index),
        colorscale='Earth',
        annotation_text=corr.round(2).values,
        showscale=True,reversescale=True)

figure.layout.margin=dict(l=200,t=200)
figure.layout.height=800
figure.layout.width=1000

iplot(figure)
```
